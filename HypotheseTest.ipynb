# <font color=brown>Bierflaschenabfüllungsproblem</font>

## Die Brauerei hat von dem Gerücht gehört. Also beschloss sie, ihre Ware zu testen zur Konfidenzniveau von 99%. Sie maß 25 Flaschen und erhielt einen Mittelwert von 224mL mit einer Standardabweichung von 9mL.

### Schritt 1)

#### $H_0$ : Die durchschnittliche Füllung beträgt 330mL ( $\mu=330$)
#### $H_1$ : Die durchschnittliche Füllung beträgt <font color=red>weniger</font> als 330 ml. ( $\mu<330$, einseitiger Test auf der linken Seite)

### Schritt 2)

### Normalverteilt, Stichprobe ist klein ($25<30$), $\sigma$ immer noch unbekannt
### $\Longrightarrow$ studentsche $t$-Verteilung mit $24$ Freiheitsgraden

### Schritt 3)

### Setzen wir <font color=red>$\alpha=0.01$</font>

### Schritt 4)

#### $\mathbb{P}(H_0\text{ ablehnen}|H_0 \text{ ist richtig}) = \alpha$ (<font color=red>Fehler 1. Art </font>)
#### $\Longleftrightarrow \mathbb{P}\left(\sqrt{n}(\overline{X}-\mu)/S_x \leq A\right)=\alpha$
#### Da $\sqrt{n}(\overline{X}-\mu)/S_x\sim T_{24}$, folgt
#### $A = -2,\!492159473$

```python
from scipy.stats import t
T24 = t(df=24)
A = T24.ppf(0.01);A
```

### Schritt 5) Wir berechnen den Wert aus unserer Stichprobe:
### $\sqrt{n}(\overline{X}-\mu)/S_X = \sqrt{25}(224-330)/9 = -10/3 = -3.33$

```python
plot_T24 = plot(T24.pdf,[-6,6],color='blue')
plot_zone1 = plot(T24.pdf,[A,6],color='green',fill=True,fillcolor='green',fillalpha=0.25)
plot_zone2 = plot(T24.pdf,[-6,A],color='blue',fill=True,fillcolor='red',fillalpha=0.75)
plot_line = line([(-10/3,0),(-10/3,0.1)],color='black',thickness=2)
tval = text(r't-Test Wert: $-10/3$', (-10/3,0.12),fontsize=12,color='black')
u_a = text(r'$1-\alpha$', (1,0.12),fontsize=12,color='black')
text_t24 = text(r'$t_{24}$', (2,0.25),fontsize=18,color='black')
bild = (plot_T24 + plot_zone1 + plot_zone2 + plot_line + tval + u_a + text_t24)
```

```python
bild.show(figsize=[9,5])
```
